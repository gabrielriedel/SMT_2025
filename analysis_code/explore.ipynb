```python
with db.connect("smt_2025.db") as con:
    df_first_runner = con.sql("""SELECT ge.game_str, gi.home_team, gi.away_team, 
                              ge.at_bat, ge.play_per_game, gi.top_bottom_inning, ge.player_position, ge.timestamp 
                              FROM game_events ge
                              LEFT JOIN 
                              LEFT JOIN game_info gi
                              ON ge.game_str = gi.game_str AND ge.play_per_game = gi.play_per_game
                              WHERE first_baserunner != 'NA'
                              LIMIT 10;""").df()
```
